## Define your first class

We are going to create a new class *person* which contains three attributes (first name, last name and age) and three methods (creation, display name and age, swap first and last names) 

```
class Person:
  
  #def __init__(self):
  #  self.firstname = "Skywalker"
  #  self.lastname = "Anakin"
  #  self.age = 42
  
  def __init__(self,firstName="Skywalker",lastName="Anakin",age=42):
    self.firstname = firstName
    self.lastname = lastName
    self.age = age
    
  def displaySentence(self):
    print('Hi, I am', self.firstname, self.lastname, 'and I am', str(self.age))
  
  def swapFirstLastName(self):
    self.firstname, self.lastname = self.lastname, self.firstname
    
```

## Else and finally

```
def divide(a,b):
  try:
    a/b
  except ZeroDivisionError:
    print('Try with a non-zero value for b')
  except TypeError:
    print('Try integer values for a and b')
  else:
    return(a/b)
```

```
a=2
b='a'
divide(a,b)
```

```
a=2
b=0
divide(a,b)
```

```
def divide(a,b):
  try:
    return(a/b)
  except ZeroDivisionError:
    print('Try with a non-zero value for b')
  except TypeError:
    print('Try integer values for a and b')
  finally:
    print('Computation complete')
```

## Assertions

```
def squareRoot(a):
  try:
    assert a>=0
  except TypeError:
    print('Try with a positive float for a')
  except AssertionError:
    print('a is negative, function shall return a complex value')
    return(a**0.5)
  else:
    return(a**0.5)    
```

## Sets

```
a = set([1, 2, 3, 4])
b = set([3, 4, 5, 6])
c = set([1, 3])
d = set([1, 2, 3 ,4 ,5, 6])
```

Intersection: common values of the two sub-sets

```
print(a.intersection(b))
print(a&b)
```

Union: values that are at least in one subset

```
print(a.union(b))
print(a|b)
```

Subset: is the set included in a superset given as argument

```
print(a.issubset(c))
print(a <= c)
print(a.issubset(d))
print(a <= d)
```

Superset: does the set includes the one given as argument 

```
print(a.issuperset(c))
print(a >= c)
print(a.issuperset(d))
print(a >= d)
```

Difference: set without the value in the set given as argument

```
print(a.difference(b))
print(a-b)
```

Symmetric Difference: complementary of the union, i.e. values that are only in one of the two sets

```
print(a.symmetric_difference(b))
print(a^b)
```

## Groups

```
X.groups()
```

Do not confuse groups and matchs. There are two matches of pattern in the string but one single group. search() returns only the first group.

```
newPattern='([A-Z]\w+).*([A-Z]\w+)'
rex = re.compile(newPattern)
X=rex.search(string)
```

```
print(X.groups())
print(X.group(0))
print(X.group(1))
print(X.group(2))
```

```
print(string[X.start(1):X.end(1)])
print(string[X.start(2):X.end(2)])
```

```
newPattern='(?P<country>[A-Z]\w+).*(?P<capital>[A-Z]\w+)'
rex = re.compile(newPattern)
X=rex.search(string)
print(X.groupdict())
```

## Sympy

```
exprstr = '(x**3 + x**2 - x - 1)/(x**2 + 2*x + 1)'
expr = sympy.sympify(exprstr)
sympy.simplify(expr)
```

```
polystr = 'a*x**2 + b*x + c'
expr = sympy.sympify(polystr)
print(expr.integrate(x))
print(expr.diff(x))
```

```
polystr = 'sin(x)/x'
expr = sympy.sympify(polystr)
print(expr.limit(x,0))
print(expr.series(x,0,10))
```

```
x, y, z = sympy.symbols("x y z")
eqn1 = sympy.sympify('3*x+4*y-5*z+6')
eqn2 = sympy.sympify('2*x+3*y-4*z+5')
eqn3 = sympy.sympify('-2*x+3*y-4*z-3')
sympy.linsolve([eqn1,eqn2,eqn3], (x, y, z))
```
